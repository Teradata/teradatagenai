<p style="font-size:18px;font-family:Arial;color:#00233C"><b>What is a Vector Store?</b></p>

<p style="font-size:16px;font-family:Arial;color:#00233C">
A vector store contains a vectorized representation of data, typically created using embeddings from AI models. It allows for high-speed similarity searches beyond traditional keyword matching.
</p>

<p style="font-size:16px;font-family:Arial;color:#00233C">
Vector embeddings are numerical representations of data (text, images, audio, etc.) mapped into a multi-dimensional space.
</p>

<p style="font-size:20px;font-family:Arial;color:#00233C"><b>Vector Search Algorithms</b></p>

<p style="font-size:16px;font-family:Arial;color:#00233C">
Teradata supports three main search algorithms:
</p>

<p style="font-size:18px;font-family:Arial;color:#00233C"><b>1. VectorDistance</b></p>
<ul style="font-size:16px;font-family:Arial;color:#00233C">
  <li><b>Flat index</b> (performs a full-table scan)</li>
  <li><b>Scalable and parallel search</b></li>
  <li><b>Exact search approach</b></li>
  <li>Suitable for <b>smaller datasets (<10K documents)</b></li>
</ul>

<p style="font-size:18px;font-family:Arial;color:#00233C"><b>2. Kmeans</b></p>
<ul style="font-size:16px;font-family:Arial;color:#00233C">
  <li><b>IVF (Inverted Vector File) index</b> (organizes data into clusters)</li>
  <li><b>Scalable and parallel training/search</b></li>
  <li><b>Approximate search approach</b></li>
  <li>Suitable for <b>larger datasets (>10K documents)</b></li>
</ul>

<p style="font-size:18px;font-family:Arial;color:#00233C"><b>3. HNSW</b></p>
<ul style="font-size:16px;font-family:Arial;color:#00233C">
  <li><b>HNSW (Hierarchical Navigable Small World) index</b></li>
  <li><b>Graph-based Approximate Nearest Neighbors (ANN) search</b></li>
  <li><b>Scalable and parallel training/search</b></li>
  <li><b>Approximate search approach</b></li>
  <li>Suitable for <b>larger datasets (>10K documents)</b></li>
</ul>

<b style = 'font-size:30px;font-family:Arial;color:#00233C'>1. Getting Started with Vector Store </b>




### 🛠 Steps to Create a Vector Store  

1. **Initialize a VectorStore Instance**  
   - Create an instance of `VectorStore`

2. **Create the Vector Store**  
   - Use the `create()` method on the vector store instance to create a new vector store.

By following these steps, we will successfully create vector store - `vs_example`.

```python
vs = VectorStore(name="vs_example", log=True)
```

```python
# Create an input DataFrame as amazon_reviews_25 is in 'oaf' database.
input_data = DataFrame(in_schema('oaf', 'amazon_reviews_25'))
```

```python
# Python
vs.create(embeddings_model= 'amazon.titan-embed-text-v1',
          search_algorithm= 'HNSW',
          seed=10,
          top_k=10,
          ef_construction=32,
          num_connpernode=32,
          maxnum_connpernode=32,
          metric='EUCLIDEAN',
          apply_heuristics=False,
          ef_search=32,
          object_names= input_data,
          key_columns= ['rev_id', 'aid'],
          data_columns= ['rev_text'],
          vector_column= 'VectorIndex')
```

### 🔍 Performing Similarity Search.

\**similarity search** allows us to find the most relevant vectors based on a given input. By passing a question to the vector store `vs_example`, we can retrieve the closest matches based on the selected search algorithm.  

#### 🛠 How It Works  

1. **Input a question**  
   - Pass a question or input text to `question` argument of `similarity_search()`.  

2. **Perform the Similarity Search**  
   - The vector store searches for the most relevant matches using the specified **`search_algorithm`**.  

3. **Retrieve Top Matches**  
   - The result contains the **top_k** most relevant entries along with their similarity scores, helping us understand how closely they match the input.  

This process enables efficient retrieval of information based on vector embeddings, making it useful for applications such as **semantic search, recommendation systems, and document retrieval**. 🚀  


```python
question = 'Which book are all the reviews talking about?'
response = vs.similarity_search(question=question)
```

```python
response.similar_objects
```

### 📝 Generating a Natural Language Response  

Once the **`similarity_search()`** method retrieves relevant results from the vector store, we can generate a natural language response by proving a prompt..

#### 🛠 Steps to Prepare the Response  
1.  **Input a question**  
   - Pass a question or input text to `question` argument of `prepare_response()`.  

1. **Retrieve Similarity Results**  
   - The `similarity_search()` method returns the most relevant matches along with their similarity scores.  

3. **Input a prompt**  
   - By passing `prompt`, we ask the model to format the output in a particular way
  
#### As we can see below, by passing the question and the prompt, the response is formated in a conversational manner as suggested in the prompt.


```python
question='Did any one feel the book is thin?'
prompt='Format the response in a conversational way.'
response = vs.prepare_response(question=question, similarity_results=response, prompt=prompt)
```

```python
print(response)
```

### ⚡ ask: 
#### Unified Similarity Search & Response Generation  
To improve efficiency, we can **combine similarity search and response generation** into a single operation. This method performs a similarity search in the vector store using the **input query** and then prepares a natural language response based on the retrieved results.  

#### 🔍 How It Works  

1.  **Input a question**  
   - Pass a question or input text to `question` argument of `ask()`.  

3. **Input a prompt**  
   - By passing `prompt`, we ask the model to format the output in a particular way

**Faster Execution**  
   - By combining both steps into a single function call, response time is reduced, making the system more efficient.  


```python
custom_prompt = """List good reviews about the books. Do not assume information. Only provide information that is present in the data.
Format results like this:
Review ID: 
Author ID: 
Review: 
"""
question ='Are there any reviews saying that the books are inspiring?'
response = vs.ask(question=question, prompt=custom_prompt)
```

```python
print(response)
```

<b style="font-size:30px; font-family:Arial; color:#00233C;">2. Initializing and Using the Existing Vector Store</b>


### 🔄 Using an Existing Vector Store  

If you have an existing vector store that you want to reuse, you can initialize it by providing the **name** of the store. This allows you to interact with and perform operations on the existing store without creating a new one.  

#### 🛠 Steps to Initialize existing Vector Store  

1. **Provide the Name of the Vector Store**  
   - Specify the name (e.g., `vs_example`) of the existing vector store that you want to access.
     

2. **Initialize the Vector Store Instance**  
   - By using the provided name, you can initialize the vector store instance and begin interacting with it.  



```python
# Initialize existing VS
updated_vs = VectorStore("vs_example")
```

### ✨ Updating Arguments of an Existing Vector Store  

You can **update** specific parameters of an existing vector store without having to recreate it. For instance, you may want to modify the **search algorithm** or other configuration settings to better suit your needs. This can be done using the `update()` API, which allows you to change the arguments of the vector store.

#### 🛠 Steps to Update an Existing Vector Store  

1. **Access the Existing Vector Store**  
   - First, ensure that the vector store (e.g., `vs_update`) is already initialized.

2. **Use the `update()` API**  
   - Call the `update()` method on the vector store instance and provide the new argument(s) you want to modify, such as the `search_algorithm`.  


```python
updated_vs.update(search_algorithm="KMEANS")
```

```python
updated_vs.status()
```

### ✨ Updating Data of an Existing Vector Store  

You can **add** or **delete** data in a Vector Store.

#### 🛠 Steps to Update an Existing Vector Store  

1. **Access the Existing Vector Store**  
   - First, ensure that the vector store (e.g., `vs_update`) is already initialized.

2. **Use the `update()` API**  
   - Call the `update()` method on the vector store instance and provide the new argument(s). 

```python
# Create a new table with the contents to add to the Vector Store. 
# Make sure this table has the columns which are used as key_columns in existing Vector Store.
import pandas as pd
from teradataml import copy_to_sql
from teradatasqlalchemy import VARCHAR
df = {'prod_name': ['A1', 'A2', 'A3', 'A4'],
      'rev_id': ['100', '200', '300', '400'],
      'rev_text': ['Product is good', 'quality is bad', 'needs improvement', 'Bad quality'],
      'aid': ['99.99', '97.32', '94.67', '91.00']
      }
pandas_df = pd.DataFrame(df)
copy_to_sql(df=pandas_df, table_name='test_table_update',
            if_exists='replace',
            types={'rev_text': VARCHAR(64000)})
```

```python
updated_vs.update(object_names='test_table_update',
                  alter_operation="ADD",
                  update_style="MINOR")
```

```python
updated_vs.status()
```

```python
op = updated_vs.get_indexes_embeddings()
```

```python
op.loc[op['TableName'] == 'test_table_update']
```

### Delete contents of 'test_table_update' from Vector Store.

```python
updated_vs.update(object_names='test_table_update',
                  alter_operation="DELETE",
                  update_style="MINOR")
```

```python
updated_vs.status()
```

```python
op = updated_vs.get_indexes_embeddings()
```

```python
op.head()
```

```python
op.loc[op['TableName'] == 'test_table_update']
```

### 🔍 Performing Similarity Search, Preparing Response, and Asking on the Updated Vector Store  

Once you've updated the vector store, you can seamlessly perform a **similarity search**, prepare a **natural language response**, and ask the updated vector store to retrieve the most relevant information.

#### 🛠 Steps to Perform on the Updated Vector Store  

1. **Similarity Search**  
   - Use the `similarity_search()` method to find the most relevant matches for your input question based on the updated configuration.


```python
question = 'Which book are all the reviews talking about?'
response = updated_vs.similarity_search(question=question)
```

```python
response.similar_objects
```

2. **Prepare the Response**  
   - After retrieving the search results, use the `prepare_response()` function to generate a user-friendly, natural language response based on the most relevant matches.


```python
question = 'Which book are all the reviews talking about?'
response = updated_vs.prepare_response(question=question, similarity_results=response)
```

```python
print(response)
```

3. **Ask the Updated Vector Store**  
   - Combine these operations using the `ask()` method, which performs the search and generates the response in one call, ensuring faster and more efficient retrieval of information.

```python
custom_prompt = """List good reviews about the books. Do not assume information. Only provide information that is present in the data.
Format results like this:
review_id: 
Book Title: 
Author: 
Review: 
"""
question="Are there any reviews saying that the books are inspiring?"
response = updated_vs.ask(question=question, prompt=custom_prompt)
```

```python
print(response)
```
