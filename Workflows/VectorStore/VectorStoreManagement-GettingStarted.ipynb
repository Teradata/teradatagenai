<p style="font-size:16px;font-family:Arial;color:#00233C">

<p style="font-size:20px;font-family:Arial;color:#00233C"><b>Vector Search Algorithms</b></p>

<p style="font-size:16px;font-family:Arial;color:#00233C">
Teradata supports three main search algorithms:
</p>

<p style="font-size:18px;font-family:Arial;color:#00233C"><b>1. VectorDistance</b></p>
<ul style="font-size:16px;font-family:Arial;color:#00233C">
  <li><b>Flat index</b> (performs a full-table scan)</li>
  <li><b>Scalable and parallel search</b></li>
  <li><b>Exact search approach</b></li>
  <li>Suitable for <b>smaller datasets (<10K documents)</b></li>
</ul>

<p style="font-size:18px;font-family:Arial;color:#00233C"><b>2. Kmeans</b></p>
<ul style="font-size:16px;font-family:Arial;color:#00233C">
  <li><b>IVF (Inverted Vector File) index</b> (organizes data into clusters)</li>
  <li><b>Scalable and parallel training/search</b></li>
  <li><b>Approximate search approach</b></li>
  <li>Suitable for <b>larger datasets (>10K documents)</b></li>
</ul>

<p style="font-size:18px;font-family:Arial;color:#00233C"><b>3. HNSW</b></p>
<ul style="font-size:16px;font-family:Arial;color:#00233C">
  <li><b>HNSW (Hierarchical Navigable Small World) index</b></li>
  <li><b>Graph-based Approximate Nearest Neighbors (ANN) search</b></li>
  <li><b>Scalable and parallel training/search</b></li>
  <li><b>Approximate search approach</b></li>
  <li>Suitable for <b>larger datasets (>10K documents)</b></li>
</ul>

<b style="font-size:30px; font-family:Arial; color:#00233C;">4.Access Management in Vector Store</b>


```python
# Initialize and create a Vectore Store
vs = VectorStore(name="acces_management_vs")
# Create an input DataFrame as amazon_reviews_25 is in 'oaf' database.
input_data = DataFrame(in_schema('oaf', 'amazon_reviews_25'))
# Python
vs.create(embeddings_model= 'amazon.titan-embed-text-v1',
          search_algorithm= 'HNSW',
          seed=10,
          top_k=10,
          ef_construction=32,
          num_connpernode=32,
          maxnum_connpernode=32,
          metric='EUCLIDEAN',
          apply_heuristics=False,
          ef_search=32,
          object_names= input_data,
          key_columns= ['rev_id', 'aid'],
          data_columns= ['rev_text'],
          vector_column= 'VectorIndex')
```

```python
vs.status()
```

You can manage both **admin** and **user** permissions for users in your vector store. This section shows how to **grant** and **revoke** these permissions to control access effectively.
**Note Only admin user can view the permissions or grant/revoke user/admin access to Vector Store.**

### 👥 Listing User Permissions in the Vector Store  

To view the permissions currently assigned to users in the vector store, you can use the `list_user_permissions()` method. This will display a list of all users and the respective access rights they have on the Vector Store.


```python
vs.list_user_permissions()
```

#### 🔒 Granting and Revoking user Permission from User 'alice'  

You can grant or revoke **USER** permission for specific users to control their access level to the vector store.
Below is how you can handle the **USER** permission for user `alice`.

##### 🛠 Steps to Grant or Revoke USER Permission for User 'alice'  

1. **Grant USER Permission**  
   - Use the `grant.user()` method to assign **USER** permission to user `alice`.

```python
vs.grant.user('alice')
```

```python
# check granted permission
vs.list_user_permissions()
```

2. **Revoke USER Permission**  
   - Use the `revoke.user()` method to remove **USER** permission from user `alice`.

```python
vs.revoke.user('alice')
```

```python
# check granted permission
vs.list_user_permissions()
```

#### 🔒 Granting and Revoking ADMIN Permission from User 'alice'  

To manage the **ADMIN** access for a user, such as `alice`, you can use the `grant.admin()` and `revoke.admin()` methods. This allows you to control who can modify the contents of the vector store.

##### 🛠 Steps to Grant or Revoke ADMIN Permission for User 'oaf_demo'  

1. **Grant ADMIN Permission**  
   - Use the `grant.admin()` method to assign **ADMIN** permission to user `oaf_demo`.

```python
vs.grant.admin('oaf_demo')
```

```python
# check granted permission
vs.list_user_permissions()
```

```python
vs.revoke.admin('oaf_demo')
```

```python
# check permission
vs.list_user_permissions()
```
