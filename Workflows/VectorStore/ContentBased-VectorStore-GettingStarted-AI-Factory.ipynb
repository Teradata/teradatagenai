<p style="font-size:18px;font-family:Arial;color:#00233C"><b>What is a Vector Store?</b></p>

<p style="font-size:16px;font-family:Arial;color:#00233C">
A vector store contains a vectorized representation of data, typically created using embeddings from AI models. It allows for high-speed similarity searches beyond traditional keyword matching.
</p>

<p style="font-size:16px;font-family:Arial;color:#00233C">
Vector embeddings are numerical representations of data (text, images, audio, etc.) mapped into a multi-dimensional space.
</p>


<p style="font-size:16px;font-family:Arial;color:#00233C">
Each embedding is a vector that captures semantic or content-based relationships. For example:

<p style="font-size:20px;font-family:Arial;color:#00233C"><b>Vector Search Algorithms</b></p>

<p style="font-size:16px;font-family:Arial;color:#00233C">
Teradata supports three main search algorithms:
</p>

<p style="font-size:18px;font-family:Arial;color:#00233C"><b>1. VectorDistance</b></p>
<ul style="font-size:16px;font-family:Arial;color:#00233C">
  <li><b>Flat index</b> (performs a full-table scan)</li>
  <li><b>Scalable and parallel search</b></li>
  <li><b>Exact search approach</b></li>
  <li>Suitable for <b>smaller datasets (<10K documents)</b></li>
</ul>

<p style="font-size:18px;font-family:Arial;color:#00233C"><b>2. Kmeans</b></p>
<ul style="font-size:16px;font-family:Arial;color:#00233C">
  <li><b>IVF (Inverted Vector File) index</b> (organizes data into clusters)</li>
  <li><b>Scalable and parallel training/search</b></li>
  <li><b>Approximate search approach</b></li>
  <li>Suitable for <b>larger datasets (>10K documents)</b></li>
</ul>

<p style="font-size:18px;font-family:Arial;color:#00233C"><b>3. HNSW</b></p>
<ul style="font-size:16px;font-family:Arial;color:#00233C">
  <li><b>HNSW (Hierarchical Navigable Small World) index</b></li>
  <li><b>Graph-based Approximate Nearest Neighbors (ANN) search</b></li>
  <li><b>Scalable and parallel training/search</b></li>
  <li><b>Approximate search approach</b></li>
  <li>Suitable for <b>larger datasets (>10K documents)</b></li>
</ul>

### ðŸ” Performing Similarity Search.

\**similarity search** allows us to find the most relevant vectors based on a given input. By passing a question to the vector store `aionprem`, we can retrieve the closest matches based on the selected search algorithm.  

#### ðŸ›  How It Works  

1. **Input a question**  
   - Pass a question or input text to `question` argument of `similarity_search()`.  

2. **Perform the Similarity Search**  
   - The vector store searches for the most relevant matches using the specified **`search_algorithm`**.  

3. **Retrieve Top Matches**  
   - The result contains the **top_k** most relevant entries along with their similarity scores, helping us understand how closely they match the input.  

This process enables efficient retrieval of information based on vector embeddings, making it useful for applications such as **semantic search, recommendation systems, and document retrieval**. ðŸš€  


```python
question = 'Which book are all the reviews talking about?'
response = vs.similarity_search(question=question)
```

```python
response.similar_objects
```

### ðŸ“ Generating a Natural Language Response  

Once the **`similarity_search()`** method retrieves relevant results from the vector store, we can generate a natural language response by proving a prompt..

#### ðŸ›  Steps to Prepare the Response  
1.  **Input a question**  
   - Pass a question or input text to `question` argument of `prepare_response()`.  

1. **Retrieve Similarity Results**  
   - The `similarity_search()` method returns the most relevant matches along with their similarity scores.  

3. **Input a prompt**  
   - By passing `prompt`, we ask the model to format the output in a particular way
  
#### As we can see below, by passing the question and the prompt, the response is formated in a conversational manner as suggested in the prompt.


```python
question='Did any one feel the book is thin?'
prompt='Format the response in a conversational way.'
response = vs.prepare_response(question=question, similarity_results=response, prompt=prompt)
```

```python
print(response)
```

### âš¡ ask:

To improve efficiency, we can **combine similarity search and response generation** into a single operation. This method performs a similarity search in the vector store using the **input query** and then prepares a natural language response based on the retrieved results.  

#### ðŸ” How It Works  

1.  **Input a question**  
   - Pass a question or input text to `question` argument of `ask()`.  

3. **Input a prompt**  
   - By passing `prompt`, we ask the model to format the output in a particular way

**Faster Execution**  
   - By combining both steps into a single function call, response time is reduced, making the system more efficient.  


```python
custom_prompt = """List good reviews about the books. Do not assume information. Only provide information that is present in the data.
Format results like this:
Review ID: 
Author ID: 
Review: 
"""
question ='Are there any reviews saying that the books are inspiring?'
response = vs.ask(question=question, prompt=custom_prompt)
```

### âœ¨ Updating Data of an Existing Vector Store  

You can **add** or **delete** data in a Vector Store.

#### ðŸ›  Steps to Update an Existing Vector Store

   - First, ensure that the vector store (e.g., `vs_update`) is already initialized.

2. **Use the `update()` API**  
   - Call the `update()` method on the vector store instance and provide the new argument(s). 

```python
# Create a new table with the contents to add to the Vector Store. 
# Make sure this table has the columns which are used as key_columns in existing Vector Store.
import pandas as pd
from teradataml import copy_to_sql
from teradatasqlalchemy import VARCHAR
df = {'prod_name': ['A1', 'A2', 'A3', 'A4'],
      'rev_id': ['100', '200', '300', '400'],
      'rev_text': ['Product is good', 'quality is bad', 'needs improvement', 'Bad quality'],
      'aid': ['99.99', '97.32', '94.67', '91.00']
      }
pandas_df = pd.DataFrame(df)
copy_to_sql(df=pandas_df, table_name='test_table_update',
            if_exists='replace',
            types={'rev_text': VARCHAR(64000)})
```

```python
updated_vs.update(object_names='test_table_update',
                  alter_operation="ADD",
                  update_style="MINOR")
```

```python
updated_vs.status()
```

```python
op = updated_vs.get_indexes_embeddings()
```

### ðŸ›  Instantiating a Content Based Vector Store using `from_texts` method

   - Creates a new 'content-based' vector store from the input text(s) or list of text(s).
   - Internally the texts are converted into table and stored to Teradata Vantage.

```python
# Sample text documents for demonstration
sample_texts = [
    "Teradata Vantage is a modern data platform that unifies data lakes, data warehouses, analytics, and new data sources.",
    "Machine learning models can be deployed directly in Teradata Vantage for in-database analytics and scoring.",
    "Vector databases enable semantic search capabilities by storing and querying high-dimensional embeddings.",
    "Natural language processing with large language models has revolutionized how we interact with data.",
    "SQL is the standard language for managing and querying relational databases across various platforms.",
    "Data science workflows often involve data preparation, model training, validation, and deployment phases.",
    "Cloud computing provides scalable infrastructure for modern data analytics and machine learning workloads.",
    "Real-time analytics enable organizations to make data-driven decisions with minimal latency."
]

print(f"Prepared {len(sample_texts)} text samples for vector store creation")
```

```python
# Create text-based vector store
try:
    vs_instance_texts = VectorStore.from_texts(name="vs_example_from_texts",
                                               texts=sample_texts,
                                               embedding = "nvidia/llama-3.2-nv-embedqa-1b-v2",
                                               embeddings_dims = 2048,
                                               embeddings_base_url = embeddings_url,
                                               completions_base_url = completions_url,
                                               chat_completion_model = "mistralai/mistral-7b-instruct-v0.3",
                                               chat_completion_max_tokens = 8192
                                               )
    
    print("Text-based vector store created successfully!")
    print(f"Vector store name: {vs_instance_texts.name}")
except Exception as e:
    print(f"Error creating text-based vector store: {e}")
```

```python
# Check the status of the store.
vs_instance_texts.status()
```

```python
# Additional texts to add
additional_texts = [
    "Artificial intelligence is transforming industries through automation and intelligent decision-making.",
    "Data governance ensures data quality, security, and compliance across enterprise systems.",
    "Business intelligence tools help organizations visualize and understand their data patterns."
]

# Add texts to existing vector store
try:
    vs_instance_texts.add_texts(texts=additional_texts)
    print(f"Successfully added {len(additional_texts)} additional texts to the vector store")
except Exception as e:
    print(f"Error adding texts: {e}")
```

```python
# Check if update is complete.
vs_instance_texts.status()
```

### ðŸ” Performing Similarity Search, Preparing Response, and Ask

Using the vector store instance created from texts, you can do **similarity search**, prepare a **natural language response**, and ask the updated vector store to retrieve the most relevant information.



```python
question = 'Which book are all the reviews talking about?'
response = vs_instance_texts.similarity_search(question=question)
```

```python
response.similar_objects
```

```python
question = 'Which book are all the reviews talking about?'
response = vs_instance_texts.prepare_response(question=question, similarity_results=response)
print(response)
```
